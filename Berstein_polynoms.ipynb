# Polynômes de Bernstein, théorème de Weierstrass

Théorème de Weierstrass :

Pour toute fonction $f(x)$ continue de $[a;b]$ dans $\mathbb K$, il existe une suite $(P_n : [a;b] \rightarrow \mathbb K)$ de polynômes convergeant uniformément vers $f$ sur $[a;b]$. 

```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Formule des polynômes de Bernstein:
$$ 
\forall x \in [0;1]: B_n(f)(x) = \sum_{k=0}^{n} . C_n^k . f(\frac{k}{n}) . x^k . (1-x)^{n-k}
$$

Avec :
$$ C_n^k = \frac{n!}{k!(n-k)!} $$

```python
def factorial(n):
    return np.prod(range(1,n+1))
        
def Ckn(k,n):
    return factorial(n) / ( factorial(k) * factorial(n-k) )
        
# attention pour n>20 ça part en sucette.
def bernstein(f,n,x):
    B = 0
    for i in range(1,n+1):
        B += Ckn(i,n) * f(i/n) * np.power(x,i) * np.power(1-x,n-i)
    return B

def fonction(x):
    return np.sin(2*pi*x)

x = np.linspace(0, 1, 100)

fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = (20, 10)
ax.plot(x, fonction(x), label='sin(2*pi*x)', color='red', 
        alpha=0.3, linestyle='solid', linewidth=2, marker='o', markersize=3,
        markerfacecolor='yellow', markeredgecolor='red')

for j in range(1,20,1):
    ax.plot(x, bernstein(fonction, j, x),
           color='black', alpha=0.2)
    
ax.plot(x, bernstein(fonction, 20, x), label='Bernstein 20', color='black')
    
ax.set_xlabel('x (rad)')
ax.set_ylabel('f(x) = sin(2*pi*x)')
ax.set_ylim([-1.1,1.1])
ax.set_title('f(x) = sin(2*pi*x)')
ax.grid(True,linestyle=':')
ax.legend()

```
